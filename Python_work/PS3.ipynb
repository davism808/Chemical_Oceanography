```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Hs = 100
Hd = 3640
Fsd = 3.5
Jp0 = 4305
Jo0 = -662970
Pd0 = 2255
Ps0 = 1025
Od0 = 92460
Os0 = 281875

Pd = [Pd0]
Ps = [Ps0]
Od = [Od0]

time = []
for t in range(0,1000000):
    time.append(t)
```

```python
for i in time:
    if Ps[i] > 0:
        Jp = Jp0
        Jo = Jo0
    if Ps[i] <= 0:
        Jp = 3.5*Pd[i]
        Jo = Jp*-154
    dPs = ((Fsd*(Pd[i]-Ps[i])-Jp)/Hs)
    dPd = (((Fsd*(Ps[i]-Pd[i]))+(0.99*Jp))/Hd)
    dOd = (((Fsd*(Os0-Od[i]))+(0.99*Jo))/Hd)
    Ps.append(dPs + Ps[i])
    Pd.append(dPd + Ps[i])
    Od.append(dOd + Od[i])
```

```python
time.append(1000001)
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(time,Ps)
ax1.plot(time,Pd)
ax2.plot(time,Od)
```
